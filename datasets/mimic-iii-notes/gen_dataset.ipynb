```python
import pandas as pd
```

```python
TRAIN_SIZE = 1000
RAW_DS_NAME = '/data/huengchi/Research/SPEER/Datasets_Processed/LLM_FT_Dataset_len8K_cov0.5_w_ESGs.jsonl'
OUTPUT_DS_NAME_PREFIX = 'NOTES_len6K_src5900_cov0.5_w_ESGs'
```

```python
ds_df = pd.read_json(RAW_DS_NAME, lines=True)
ds_df.columns
```

```python
df_to_save = ds_df[
    ds_df.apply(
        lambda row: row['len_total_tok']<6144,
        axis=1
    )
]
df_to_save
```

```python
df_to_save.sort_values(by='len_src_tok')
```

```python
# full
df_to_save.to_json(f'{OUTPUT_DS_NAME_PREFIX}.jsonl', lines=True, orient='records')
```

```python
# train_split
df_to_save[:TRAIN_SIZE].to_json(f'{OUTPUT_DS_NAME_PREFIX}_train_split.jsonl', lines=True, orient='records')
```

```python
# test_split
df_to_save[TRAIN_SIZE:].to_json(f'{OUTPUT_DS_NAME_PREFIX}_test_split.jsonl', lines=True, orient='records')
```
